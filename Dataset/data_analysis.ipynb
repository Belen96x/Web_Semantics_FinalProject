# USA dataset

```python
df_usa[['date', 'time']] = df_usa['date_time'].str.split(' ', expand=True)
```

```python
df_usa = df_usa.fillna('unknown')
```

```python
#Extracting witnesses information
occupations = [
    'unknown', 'Air Traffic Control employee', 'Police Officer',
    'MoD Guard Service', 'Former Air Traffic Controller', 'Pilot',
    'Air Traffic Controller', 'Ex-test engineer for Hawker Siddley',
    'Part-Time Pilot', "Has pilot's licence", 'Ex-police',
    'Professional Photographer', 'Air Traffic Control', 'Police',
    'Retired merchant seaman', 'Retired RAF', 'Amateur astronomer',
    'RN Air Traffic Control', 'Ex-pilot', 'Journalist', 'Ex-RAF',
    'RAF', 'Police Sgt', 'Ex-RN Cdr'
]

# Normalize occupations to ignore \n, casing, and symbols
normalized_occupations = [re.sub(r'[\s\n]', '', occupation).lower() for occupation in occupations]

# Function to find the occupation in the description
def find_occupation(description):
    if pd.isna(description):
        return 'unknown'
    normalized_description = re.sub(r'[\s\n]', '', description).lower()
    for occupation, norm_occupation in zip(occupations, normalized_occupations):
        if norm_occupation in normalized_description:
            return occupation
    return 'unknown'

# Apply the function to create the 'occupation' column
df_usa['occupation'] = df_usa['description'].apply(find_occupation)

# Display the modified DataFrame
df_usa.head()
```

# UK dataset

```python
df_uk = pd.read_csv(fr'C:\Users\belen\Desktop\Université de Lorraine\Second semester\Web_semantics_FinalProject\Web_Semantics_FinalProject\Dataset\UK_UFO_sightings.csv', sep=';',encoding='cp1252')
```

```python
df_uk['Date'] = df_uk['Date'].str.replace('-', '/')
```

```python
#Clean the rows that dont follow the wanted format

pattern = '[A-Za-z]+/[0-9]{2}'

mask = ~df_uk['Date'].str.contains(pattern)
filtered_df = df_uk[mask]
df_uk = filtered_df

df_uk.head()
```

```python
# Rename the columns to lowercase and replace spaces with underscores
df_uk.columns = df_uk.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop the last unnamed column
df_uk = df_uk.loc[:, ~df_uk.columns.str.contains('^unnamed')]

df_uk = df_uk.rename(columns={
    'occupation\n(where_relevant)': 'occupation',
    'town_/_village': 'town_village'
})

#normalizing NaN data

df_uk = df_uk.fillna('unknown')
df_uk = df_uk.replace('Not Given', 'unknown')

# Display the modified DataFrame
df_uk.head()
```

```python
#Cleaning process to extract ufo shapes 

ufo_shapes = ['cylinder', 'light', 'circle', 'sphere', 'disk', 'fireball', 'unknown', 'oval', 'other', 'cigar', 'rectangle', 'chevron', 'triangle', 'formation', 'delta', 'changing', 'egg', 'diamond', 'flash', 'teardrop', 'cone', 'cross', 'pyramid', 'round', 'crescent', 'flare', 'hexagon', 'dome', 'changed']
# Convert NaN in the list to string 'nan' for comparison
ufo_shapes = [str(shape) if pd.isna(shape) else shape for shape in ufo_shapes]

# Function to find the ufo shape in the description
def find_ufo_shape(description):
    if pd.isna(description):
        return 'unknown'
    for shape in ufo_shapes:
        if shape in description.lower():
            return shape
    return 'unknown'

# Apply the function to create the 'ufoShape' column
df_uk['ufoShape'] = df_uk['description'].apply(find_ufo_shape)

df_uk.head()
```
